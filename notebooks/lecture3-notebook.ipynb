```python
even=0
odd=0
```

```python
value=random.randint(1000,9999)
if value % 2 == 1:
    print("value is odd")
    odd = odd +1
else:
    even += 1
print("value is ", value)
print("odd ", odd)
print("even ", even)
```

```python
count=0
while count < 50:
    value=random.randint(1000,9999)
    if value % 2 == 1:
        print("value is odd")
        odd = odd +1
    else:
        even += 1
    print("value is ", value)
    print("odd ", odd)
    print("even ", even)
    count += 1
```
